# Pymaceuticals Inc.
---

The analysis below was conducted on experimental data detailing the performance of an array of drugs and their effectiveness in combating squamous cell carcinoma (SCC). Mice with SCC were identified, and tumor development was logged over a 45-day span. The objective was to compare the efficacy of Pymaceuticals' treatment, Capomulin, against that of nine other drug regimens. For each drug regimen, the mean, median, variance, standard deviation, and standard error of the tumor volume were calculated. Bar charts were generated to display the total number of entries for each drug regimen. Additionally, a pie chart was employed to illustrate the ratio of male to female mice in the experiment.

Further analysis focused on the four most promising treatment plans: Capomulin, Ramicane, Infubinol, and Ceftamin. For each of these regimens, the tumor volume after 45 days in mice was used to create box plots and identify outlier values. Additionally, a specific mouse treated with Capomulin (l509) was selected, and the tumor volume was plotted against the number of days. The average tumor volume of all mice treated with Capomulin was also plotted against their weight. The correlation coefficient was calculated and displayed.

Based on this analysis, the following conclusions were drawn:

1) Tumor progression in the selected mouse (l509) treated with Capomulin showed a relatively steady decline in tumor volume after approximately 20 days, lasting for about 15 days. Further testing is needed to determine whether the tumor volume stabilizes, increases, or remains constant after the 45-day test period.

2) A significant correlation was observed between the weight of the mice and the average tumor volume for those treated with Capomulin. It is crucial to note that no changes in the weight of the mice were recorded during the testing period. Tracking potential weight fluctuations could be essential when assessing the efficacy of Capomulin as a treatment.
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
merged_data = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
merged_data.head()
```

```python
# Checking the number of mice.
merged_data['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicates = merged_data.duplicated(subset= ['Timepoint' ,'Mouse ID'], keep=False)
extracted = merged_data.loc[duplicates, 'Mouse ID'].unique()
extracted
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = merged_data[merged_data['Mouse ID'] == 'g989']
duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_data = merged_data[merged_data['Mouse ID'] != 'g989']
cleaned_data.head()
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_data['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
data_mean = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
data_median = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
data_variance = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
data_stdev = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
data_sem = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats = pd.DataFrame({
    "Mean Tumor Volume": data_mean,
    "Median Tumor Volume": data_median,
    "Tumor Volume Variance": data_variance,
    "Tumor Volume Std. Dev.": data_stdev,
    "Tumor Volume Std. Err.": data_sem
})

summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
summary_statistics = cleaned_data.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
# Using the aggregation method, produce the same summary statistics in a single line
summary_statistics
```

## Quartiles, Outliers and Boxplots

```python
cleaned_data.head()
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
max_timepoint = cleaned_data.groupby('Mouse ID')['Timepoint'].max()
max_timepoint = max_timepoint.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
timepoint_merged = max_timepoint.merge(cleaned_data,on=['Mouse ID','Timepoint'],how="left")

selected_regimens = timepoint_merged[timepoint_merged['Drug Regimen'].isin(['Capomulin','Ramicane', 'Infubinol',
                                                                           'Ceftamin'])]
selected_regimens.reset_index(drop=True, inplace=True)

selected_regimens.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    selected_drug = selected_regimens[selected_regimens['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    
    # add subset 
    tumor_vol.append(selected_drug)

    
    # Determine outliers using upper and lower bounds
    quartiles = selected_drug.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr) 
    outliers = selected_drug.loc[(selected_drug < lower_bound) | (selected_drug > upper_bound)]
    print(f"For {treatment} the IQR is {iqr}")
    if not outliers.empty:
        print(f"{treatment}'s' outliers are {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.set_xticklabels(treatments)
ax1.boxplot(tumor_vol, flierprops = dict(marker = "o", markerfacecolor = "red", markersize=11))
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
mouse_data = cleaned_data[cleaned_data['Mouse ID'] == 'l509']
mouse_data.plot('Timepoint','Tumor Volume (mm3)')
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Capomulin treatment of mouse l509");
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_data = cleaned_data[cleaned_data['Drug Regimen'] == 'Capomulin']
capomulin_grouped = capomulin_data.groupby('Mouse ID')['Tumor Volume (mm3)'].mean()
weights = capomulin_data.groupby('Mouse ID')['Weight (g)'].first()
plt.scatter(weights, capomulin_grouped, marker='o')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.title('Average Tumor Volume vs Weight for Capomulin Mice');
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(weights, capomulin_grouped)
regress_values = weights * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(weights, capomulin_grouped, marker='o')
plt.plot(weights,regress_values,"r-")
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
correlation = st.pearsonr(weights, capomulin_grouped)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
```
